```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the dfs
ts = pd.read_csv('timeseries.csv.gz')
ts.loc[:, 'ts'] = pd.to_datetime(ts['ts'])
minv = pd.read_csv('media_investments.csv.gz')
minv.loc[:, 'ts'] = pd.to_datetime(minv['ts'])
```

```python
ts['ts'].dt.date
```

```python
to_plot = ts.loc[ts['ts'].dt.date == pd.Timestamp('2020-07-05')]
minv_plot = minv.loc[minv['ts'].dt.date == pd.Timestamp('2020-07-05')]
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(to_plot['ts'], to_plot['traffic'])
for mi in minv_plot['ts'].values:
    ax.axvline(mi, linestyle='--', color='k')
```
